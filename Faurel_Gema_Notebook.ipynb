```python
import json

def explore_json_structure(data, parent_key='', depth=0):
    if isinstance(data, dict):
        for key, value in data.items():
            explore_json_structure(value, parent_key + '.' + key if parent_key else key, depth + 1)
    elif isinstance(data, list):
        for i, item in enumerate(data):
            explore_json_structure(item, parent_key + '[{}]'.format(i), depth + 1)
    else:
        print('{}: {}'.format(parent_key, type(data).__name__))

# Ubah 'nama_file.json' menjadi nama file JSON Anda
with open('/content/source/1000000054038.json') as f:
    json_data = json.load(f)

explore_json_structure(json_data)
```

```python
import os
import json
import pandas as pd

def read_json_files_in_folder(folder_path):
    json_data = []
    for filename in os.listdir(folder_path):
        if filename.endswith('.json'):
            with open(os.path.join(folder_path, filename)) as file:
                data = json.load(file)
                json_data.append(data)
    return json_data

def transform_json_to_dataframe(json_data):
    rows = []
    for data in json_data:
        sku_name = data['item_group_name']
        for sku in data['product_skus']:
            barcode = sku['item_code']
            for variation in sku['variation_values']:
                size = variation['value']
                sell_price_dict = {price['channel_id']: price['sell_price'] for price in sku['prices']}
                # Mapping channel_id to channel name
                channel_map = {
                    4: 'Lazada',
                    8: 'Zalora',
                    32: 'Blibli',
                    64: 'Shopee',
                    128: 'Tokopedia',
                    131076: 'Tiktok',
                    1048576: 'Shopify'
                }
                rows.append({
                    'Barcode': barcode,
                    'Size': size,
                    'SKU': sku_name,
                    'Lazada': sell_price_dict.get(4, 0),
                    'Zalora': sell_price_dict.get(8, 0),
                    'Blibli': sell_price_dict.get(32, 0),
                    'Shopee': sell_price_dict.get(64, 0),
                    'Tokopedia': sell_price_dict.get(128, 0),
                    'Tiktok': sell_price_dict.get(131076, 0),
                    'Shopify': sell_price_dict.get(1048576, 0)
                })
    return pd.DataFrame(rows)

folder_path = '/content/source'

json_data = read_json_files_in_folder(folder_path)

df = transform_json_to_dataframe(json_data)

df
```

```python
import pandas as pd

transaction_df = pd.read_excel("/content/process.xlsx", sheet_name="transaction")
master_process_df = pd.read_excel("/content/process.xlsx", sheet_name="masterprocess")


merged_df = pd.merge(transaction_df, master_process_df, on="process", how="left")



def check_process_order(processes):
    orders = []
    process_list = list(processes)
    for process in process_list:
        order = master_process_df[master_process_df["process"]==process]["process order"].iloc[0] if not master_process_df[master_process_df["process"]==process].empty else float('inf')
        orders.append(order)
    return sorted(process_list, key=lambda x: orders[process_list.index(x)])

result_df = merged_df.groupby("finish_good").agg({
    "process": lambda x: list(x),
    "process order": lambda x: check_process_order(x),
}).reset_index()

print(result_df)
```
